```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import math
print("Libraries imported")
```

```python
country = pd.read_csv('../data/count.csv')
col = pd.read_csv('../data/col.csv')
dft = pd.read_csv('../data/dft.csv')
decades = pd.read_csv('../data/decades.csv')
```

```python
country.drop('Unnamed: 0', axis=1, inplace=True)
dft.drop('Unnamed: 0', axis=1, inplace=True)
decades.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
def round_dataframe_to_integers(dataframe):
    rounded_data = dataframe.copy()
    for col in dataframe.columns[1:]:
        rounded_data[col] = dataframe[col].round(0).astype(int)
    return rounded_data
df= round_dataframe_to_integers(dft)
country = round_dataframe_to_integers(country)
```

```python
def lineplot_subplots(dataframe):
    x_column = dataframe.columns[0]  # Use the first column as x-values
    for i in range(1, len(dataframe.columns)):
        y_column = dataframe.columns[i]
        plt.figure(figsize=(15,5))
        plt.plot(dataframe[x_column], dataframe[y_column])
        plt.title(f"{y_column} - Line Chart")
        plt.xlabel(x_column)
        plt.ylabel(y_column)
        plt.xticks(rotation=45)
        plt.grid()
        plt.show()

```

```python
def lineplot_top_subplots(dataframe):
    x_column = dataframe.columns[0]  # Use the first column as x-values
    for i in range(1, len(dataframe.columns)):
        y_column = dataframe.columns[i]
        
        # Sort the column in descending order and take the top 10 values
        sorted_data = dataframe.sort_values(by=y_column, ascending=False).head(10)
        
        plt.figure(figsize=(8, 4))
        plt.plot(sorted_data[x_column], sorted_data[y_column])
        plt.title(f"Top 10 {y_column} - Line Chart")
        plt.xlabel(x_column)
        plt.ylabel(y_column)
        
        # Rotate the x-labels by 45 degrees
        plt.xticks(rotation=45)
        
        plt.grid()
        plt.show()
```

```python
def lineplot_bottom_subplots(dataframe):
    x_column = dataframe.columns[0]  # Use the first column as x-values
    for i in range(1, len(dataframe.columns)):
        y_column = dataframe.columns[i]
        
        # Sort the column in descending order and take the top 10 values
        sorted_data = dataframe.sort_values(by=y_column, ascending=False).tail(10)
        
        plt.figure(figsize=(8, 4))
        plt.plot(sorted_data[x_column], sorted_data[y_column])
        plt.title(f"Bottom 10 {y_column} - Line Chart")
        plt.xlabel(x_column)
        plt.ylabel(y_column)
        
        # Rotate the x-labels by 45 degrees
        plt.xticks(rotation=45)
        
        plt.grid()
        plt.show()
```

```python
def barplot_subplots(dataframe):
    x_column = dataframe.columns[0]  # Use the first column as labels
    for i in range(1, len(dataframe.columns)):
        y_column = dataframe.columns[i]
        plt.figure(figsize=(6, 6))
        plt.bar(dataframe[y_column], labels=dataframe[x_column], autopct='%1.1f%%')
        plt.title(f"{y_column} - Pie Chart")
        plt.show()
```

```python
def pieplot_subplots(dataframe):
    x_column = dataframe.columns[0]  # Use the first column as labels
    for i in range(1, len(dataframe.columns)):
        y_column = dataframe.columns[i]
        plt.figure(figsize=(6, 6))
        plt.pie(dataframe[y_column], labels=dataframe[x_column], autopct='%1.1f%%')
        plt.title(f"{y_column} - Pie Chart")
        plt.show()
```
